```python
import matplotlib.pyplot as plt
import numpy as np

def lineer_regresyon(x):
    a = 2
    b = 4
    y = a * x + b
    return y
"""
This function calculates the y value for a given x value using a linear regression model.
"""
x = np.linspace(0, 50, 10)
y = lineer_regresyon(x)
plt.plot(x, y)
plt.show()
```
